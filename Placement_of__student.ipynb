Data Preprocessing /Data Cleaning

```python
df1["Aptitude_Score"].fillna(df1["Aptitude_Score"].mean(),inplace=True) # all null values will have average value
```

```python
df1["Communication_Score"].fillna(df1["Communication_Score"].mean(),inplace=True) #all null values in communication score will get mean value
```

```python
df1["CGPA"].fillna(df1["CGPA"].mean(),inplace=True) # it will add mean cgpa in all null values present in column of cgpa
```

```python
df1["Internships"]=df1["Internships"].fillna(0) # it will fill all null values or missing values with zero
```

```python
df1["Projects"]=df1["Projects"].fillna(1) # it will add one to all null values or missing values present in  project column.
```

```python
df1["Attendance"]=df1["Attendance"].fillna(df1["Attendance"].median())# we use median to avoid outliers
```

```python
df1.loc[df1["Mock_Status"] == "Absent", "Mock_Score"] = 0
```

```python
df1["Mock_Status"]=df1["Mock_Status"].map({"Present":1,"Absent":0})
```

```python
df1["Placed"]=df1["Placed"].map({"Placed":1,"Not Placed":0})
```

```python
df1
```

```python
df1.drop_duplicates() # we drop all duplicates row
```

```python
df1.value_counts("Placed") #we will calculate the no. students placed (1)or not placed(0)
```

```python
df1.isnull().sum()
```

```python
Placed_avg_student=df1.groupby("Department")["Placed"].mean()
print(Placed_avg_student*100)
```

```python
counts=df1.groupby("Department")["Placed"].value_counts()
print(counts)
```

```python
df1["Profile_Score"]=df1["Internships"]+df1["Projects"]
```

```python
df1["Skill_Score"]=(df1["Aptitude_Score"]+df1["Communication_Score"])/2
```

```python
df1.drop(["Enrollment_ID","Name","Email"],axis=1,inplace=True)
```

```python
df1.groupby("Placed")[["CGPA","Aptitude_Score","Communication_Score"]].mean()
```

```python
df1["Placed"] = (
    (df1["CGPA"] > 7) &
    (df1["Aptitude_Score"] > 60) &
    (df1["Communication_Score"] > 60)
).astype(int) # here we set some condition to check that studets placed or not placed to improve the accuracy of model
```

Name,Email,Enrollment_Id are not much relevant to our prediction so we can skip that part in email .



The data is been cleaned as there are no null values present in the details which are required for prediction .

Model Train and Testing

```python
x=np.array(df1[["Mock_Score","CGPA","Profile_Score","Skill_Score"]])
y=np.array(df1[["Placed"]])
```

 Below we have trained the model :



1.   Logistic Regression
2.   Decission Tree
3.   Naive Bayes
4.   Support Vector Machine


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
lr_model.fit(x_train,y_train)
```

```python
lr_pred=lr_model.predict(x_test)
print("logisitic Regression \n")
print("accuracy:",accuracy_score(y_test,lr_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
```

```python
dt_pred=dt_model.predict(x_test)
print("Decision Tree Classifer \n")
print("accuracy:",accuracy_score(y_test,dt_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB
nb_model=MultinomialNB()
nb_model.fit(x_train,y_train)
```

```python
nb_predict=nb_model.predict(x_test)
print("Naive Bayes")
print("accuracy",accuracy_score(y_test,nb_predict))
```

```python
from sklearn.svm import SVC
svm_model=SVC()
svm_model.fit(x_train,y_train)
```

```python
svm_predict=svm_model.predict(x_test)
print("Support Vector Machine")
print("accuracy",accuracy_score(y_test,svm_predict))
```
